```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting only string (object) type columns from the DataFrame
str_cols = df.select_dtypes(include=['object'])

# Define the number of subplots needed, based on the number of object-type columns
num_str_cols = len(str_cols.columns)
rows = (num_str_cols + 1) // 2  # Arrange plots in two columns

# Create a figure with subplots
plt.figure(figsize=(14, rows * 5))  # Adjust size as needed

# Loop through the string columns and create a countplot for each one
for index, column in enumerate(str_cols.columns):
    plt.subplot(rows, 2, index + 1)  # Arrange subplots
    sns.countplot(y=column, data=df, palette='Set3')  # Using countplot with y=column to fit long labels
    plt.title(f'Countplot of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Convert strings/object to numerical

```python
sex_map = {"female": 0, "male": 1}
embark_map = {"C": 0, "S": 1, "Q": 2}
def numerize(df):
    if df['Sex'].dtype not in ['int', 'float']:
        df.Sex = df.Sex.map(sex_map)
    if df['Embarked'].dtype not in ['int', 'float']:
        df['Embarked'] = df.Embarked.map(embark_map)
numerize(df)

```

```python
from sklearn.preprocessing import LabelEncoder

# Creating an instance of labelencoder
labelencoder = LabelEncoder()

# Assuming df is your DataFrame and 'Category' is your column of string data
df['Name'] = labelencoder.fit_transform(df['Name'])
```

### Correlation heatmap of numeric features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['int', 'float'])
numeric_df = numeric_df[numeric_df.columns.difference(['Survived'])]  # More concise way to exclude

# Move 'Survived' column to the end
numeric_df = pd.concat([numeric_df, df[['Survived']]], axis=1)  # Concatenate with 'Survived'

# Calculate correlation matrix
corr_matrix = numeric_df.corr()

# Create heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation heatmap of numeric features")
plt.show()
```
